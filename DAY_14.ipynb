#**DICTIONARIES**

```python
#creating a dictionary
student={"name":"Akshitha","age":20,"branch":"cse"}
```

```python
#accessing a dictionary
print(student)
print(student["name"])
print(student["age"])
print(student["branch"])
```

```python
#add new key-value pair
student["year"]=2023
print(student)
```

```python
student["branch"]="ece"
print(student)
```

```python
#delete elements
d = {"a": 1, "b": 2, "c": 3}
del d["a"]
print(d)
```

```python
#loop through dictionary
marks = {"math": 90, "science": 85, "english": 92}
for key in marks:
    print(key)
for value in marks.values():
    print(value)
for key,value in marks.items():
  print(key,value)
```

```python
#count word frequency
sentence = "python is easy and python is powerful"
words=sentence.split()
freq={}
for w in words:
  if w in freq:
    freq[w]+=1
  else:
    freq[w]=1
print(freq)

#other way
from collections import Counter
print(Counter(sentence.split()))


```

```python
#reverse lookup
grades = {"Aki": 90, "Siri": 88, "Laya": 76}
search_value=88
for name,grade in grades.items():
  if grade==search_value:
    print("for the search_value",search_value,"the name is",name)
```

```python
#dictionary from two lists
keys = ["name", "age", "city"]
values = ["Aki", 20, "Hyderabad"]

result = {}

for i in range(len(keys)):
    result[keys[i]] = values[i]

print(result)
```

```python
#merge two dictnories
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}

d1.update(d2)   # merges d2 into d1

print(d1)
```

```python
#remove duplicates
data = {"a": 10, "b": 20, "c": 10, "d": 30}

# Step 1: Count frequency of each value
value_count = {}

for v in data.values():
    value_count[v] = value_count.get(v, 0) + 1

# Step 2: Create new dict without duplicate values
result = {}

for key, value in data.items():
    if value_count[value] == 1:   # keep only unique values
        result[key] = value

print(result)
```
