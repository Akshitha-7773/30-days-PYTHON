## *IF ELSE*

```python
num=int(input("enter a number:"))
if num>0:
  print("positive")
else:
  print("negative")
```

```python
age=int(input("enter your age:"))
if age>=18:
  print("adult")
else:
  print("you are a minor")
```

```python
num=int(input("enter a number:"))
if num%2==0:
  print("the number is even")
else:
  print("the number is odd")
```

```python
score=int(input("Enter the score:"))
if score>=90:
  print("O")
elif score>=80:
  print("A+")
elif score>=60:
  print("A")
else:
  print("Your grade is P ,there is no fail")
```

```python
fruits=["cherry","apple","banana"]
search=str(input("enter a fruit name:"))
if search in fruits:
  print("fruit is found")
else:
  print("fruit is not found")
```
