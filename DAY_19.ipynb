```python
#creating a dataframe

import pandas as pd
data={
    "name":["aki","riya","sam"],
    "age":[20,34,12],
    "branch":["cse","it","cse"]
    }
df=pd.DataFrame(data)
print(df)
df.to_csv("students.csv", index=False)
```

```python
# accessing columns & rows
print(df["name"])
print(df.iloc[0])
```

```python
# basic info
print(df.head())
print(df.shape)
print(df.columns)
```

```python
#read a csv file
df=pd.read_csv("students.csv")
print(df)
```

```python
# filtering data
print(df[df["age"]>20])
```

```python
#add new column
df["grade"]=["a", None,"c"]
print(df)
```

```python
# handle missing values
df.fillna("unknown",inplace=True)
print(df)
```

```python
#group by
print(df.groupby("branch"))
```

```python
# sorting data
print(df.sort_values(by="age",ascending=False))
```
