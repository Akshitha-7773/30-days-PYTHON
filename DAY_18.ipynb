```python
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr)
print(type(arr))
```

```python
import numpy as np
zeros=np.zeros(5)
ones=np.ones(5)
print("zeros:",zeros)
print("ones:",ones)
```

```python
import numpy as np
a=np.array([10,20,30])
b=np.array([1,2,3])
print(a+b)
print(a-b)
print(a*b)
```

```python
import numpy as np
arr=np.array([10,20,30,40,50])
print(arr[0])
print(arr[1:4])
print(arr[-1])
```

```python
import numpy as np
matrix=np.array([[1,2,3],
                [4,5,6]])
print(matrix)
print("shape",matrix.shape)
print("element",matrix[1][2])
```

```python
import numpy as np
arr=np.array([1,4,9,16])
print("square root:",np.sqrt(arr))
print("sum:",np.sum(arr))
print("mean:",np.mean(arr))
print("max:",np.max(arr))
```

```python
import numpy as np
marks=np.array([45,67,89,34,90])
passed=marks[marks>=50]
print("passed students marks:",passed)
```

```python
import numpy as np
random_numbers=np.random.randint(1,100,size=10)
print(random_numbers)
```

```python
import numpy as np
prices=np.array([100,200,300])
discount=10
final_price=prices-(prices*discount/100)
print(final_price)
total_bill=sum(final_price)
print(total_bill)
```
