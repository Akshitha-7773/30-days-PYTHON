Let's say we have a list L1=[1,2,2,3,2,4,5],now identify the most repetitive number in the list??

```python
num_counter={}
l1=[1,2,2,3,2,5,5]
for i in l1:
   if i in num_counter:
    num_counter[i]+=1
   else:
      num_counter[i]=1
print(num_counter)
temp=0
for k,v in num_counter.items():
  if v>temp:
    temp=v
    final_result=k
print(final_result)


```

```python
 #or
from collections import Counter

l1 = [1, 2, 2, 3, 2, 5, 5]
result = Counter(l1).most_common(1)[0][0]
print(result)
```
