```python
def even_odd(q):

  if (q%2)==0:
    return("even")
  else:
      return("odd")

result=even_odd(23)
print(result)
```

```python


def calculator(x,y,op):
  if op =="+":
    return(x+y)
  elif op=="-":
    return(x-y)
  elif op=="*":
    return(x*y)
  elif op=="/":
    return(x/y)
  else:
    print("invalid operation")

result=(calculator(10,10,"+"))
print(result)


```

```python
def square(n):
  return(n*n)
result=square(2)
print(result)
```

```python
def pos_neg(n):
  if n>=0:
    return ("positive")
  else:
    return("negative")
result=pos_neg(10)
print(result)
```

```python
def sum_list(nums):
  total=0
  for num in nums:
    total+=num
  return total
nums=sum_list([1,2,3,4,5,6])
print(nums)
```

```python
def count_even(nums):
  count=0
  for num in nums:
    if num%2==0:
     count+=1
  return count
nums=count_even([1,2,3,4,5,6,7])
print(nums)
```
