#**FILE HANDLING**

```python
with open('my_file.txt', 'w') as f:
    f.write('This is some content for the text file.')
print('File my_file.txt created successfully.')
```

```python
with open('my_file.txt', 'r') as f:
    content = f.read()
print(content)
```

```python
with open('my_file.txt', 'a') as f:
    f.write('\nThis is some additional content.')
print('Content appended to my_file.txt successfully.')
```

```python
import csv

data = [
    ['Name', 'Age'],
    ['Alice', 30],
    ['Bob', 24]
]

with open('my_data.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)

print('CSV file my_data.csv created successfully.')
```

```python
import pandas as pd

df = pd.read_csv('my_data.csv')
display(df)
```
