```python
numbers = [1, 2, 3]

try:
    print(numbers[5])
except IndexError:
    print(" IndexError: Index does not exist in the list")
```

```python
student = {"name": "Aki", "age": 19}

try:
    print(student["grade"])
except KeyError:
    print(" KeyError: Key 'grade' not found")
```
