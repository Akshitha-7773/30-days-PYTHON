#**SETS**

```python
# Creating a set
fruits = {"apple", "banana", "cherry"}
print("Fruits set:", fruits)

# Duplicates are ignored
numbers = {1, 2, 2, 3, 4, 4}
print("Numbers set (duplicates removed):", numbers)
```

```python
colors = {"red", "green", "blue"}

# Adding an element
colors.add("yellow")
print("After adding yellow:", colors)

# Removing an element
colors.remove("green")  # throws error if not found
print("After removing green:", colors)

# Safe removal
colors.discard("purple")  # no error even if not present
print("After discarding purple:", colors)
```

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

# Union
print("Union:", a | b)

# Intersection
print("Intersection:", a & b)

# Difference
print("Difference a-b:", a - b)
```
