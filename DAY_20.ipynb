```python
#simple line plot
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

```python
# add labels and title
plt.plot(x,y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("simple line graph")
plt.show()
```

```python
#simple bar chart
names=["aki","riya","sam"]
marks=[20,30,40]
plt.bar(names,marks)
plt.title("marks comparision")
plt.show()
```

```python
#plot using pandas
import pandas as pd
data={
    "students":["aki","riya","sam"],
    "marks":[20,30,40]
}
df=pd.DataFrame(data)
plt.bar(df["students"],df["marks"])
plt.show()
```

```python
# multiple lines
x=[1,2,3,4,5]
y1=[20,30,40,50,60]
y2=[15,25,35,45,55]
plt.plot(x,y1,label="line 1")
plt.plot(x,y2,label="line 2")
plt.legend()
plt.show()
```

```python
#scatter plot
age=[20,21,22,23]
marks=[70,75,80,90]
plt.scatter(age,marks)
plt.xlabel("age")
plt.ylabel("marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
ips=["1p1","ip2","ip3"]
attempts=[120,13,1]
plt.bar(ips,attempts)
plt.title("login attempts per ip")
plt.show()
```

```python
#histogram
marks=[35,40,55,60,75,80,90]
plt.hist(marks)
plt.title("marrks distribution")
plt.show()
```

```python
logs = {
    "IP": ["A", "A", "A", "B", "C", "B"]
}

df = pd.DataFrame(logs)

df["IP"].value_counts().plot(kind="bar")
plt.show()
```
